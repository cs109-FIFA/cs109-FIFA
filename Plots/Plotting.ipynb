```python
def traditional_stat_plot(features,year, top = 20,player_name = None):
    #Setting up the data:
    goals_assists = features.iloc[[5,12],:]
    goals_assists = goals_assists.append(goals_assists.iloc[0,:]+goals_assists.iloc[1,:],ignore_index=True)
    tot = np.array(goals_assists.iloc[2,:].values.astype(int))
    #Getting the top X players:
    top_ = np.argsort(tot)[-top:][::-1]
    df_top_ = goals_assists.iloc[:,top_]
    
    if player_name is None:
        top_players = [w.encode('raw_unicode_escape').decode('utf-8') for w in df_top_.columns.values]
    else:
        top_players = player_name
    #Plotting:
    sns.set_style("whitegrid")
    sns.set_context({"figure.figsize": (24, 10)})
    
    sns.barplot(x = top_players[::-1], y = df_top_.iloc[2,:][::-1], color = sns.diverging_palette(10, 220, sep=80, n=7)[6],alpha = 0.5)
    p1 = sns.barplot(x = top_players[::-1], y = df_top_.iloc[1,:][::-1], color = sns.diverging_palette(10, 220, sep=80, n=7)[0],alpha = 0.5)
    p1.set_xticklabels(top_players[::-1],rotation=60)
    p1.set_title("%s-%s Top %s players based on traditional statistics" % (year,year+1,top),fontsize = 25)
    sns.despine(left=True)
    p1.set(ylabel="")
    topbar = plt.Rectangle((0,0),1,1,fc = sns.diverging_palette(10, 220, sep=80, n=7)[6],alpha = 0.5, edgecolor = 'none')
    bottombar = plt.Rectangle((0,0),1,1,fc=sns.diverging_palette(10, 220, sep=80, n=7)[0],alpha = 0.5,  edgecolor = 'none')
    p1.legend([topbar,bottombar], ['Assists','Goals'],loc=2, ncol = 1, prop={'size':20})
    for i in range(top):
        p1.text(-0.16+i, df_top_.iloc[1,(top-1)-i]/2-0.5,str(df_top_.iloc[1,(top-1)-i]), fontsize = 20,color = 'white')
        if df_top_.iloc[0,(top-1)-i] >2:
            if df_top_.iloc[0,(top-1)-i]<10:
                p1.text(-0.10+i, df_top_.iloc[1,(top-1)-i]-0.5 + df_top_.iloc[0,(top-1)-i]/2,str(df_top_.iloc[0,(top-1)-i]), fontsize = 20,color = 'white')
            else:
                p1.text(-0.16+i, df_top_.iloc[1,(top-1)-i]-0.5 + df_top_.iloc[0,(top-1)-i]/2,str(df_top_.iloc[0,(top-1)-i]), fontsize = 20,color = 'white')
    return p1
```

```python
def metric_plot(features,year,top = 20,player_name = None):
    
    metric_ = features.iloc[23,:]
    top_ = np.argsort(metric_.values)[-top:][::-1]
    df_top_ = metric_[top_]
    
    if player_name is None:
        top_players = [w.encode('raw_unicode_escape').decode('utf-8') for w in df_top_.index.values]
    else:
        top_players = player_name
        
    floored = [int(np.floor(d)) for d in df_top_]

    sns.set_style("whitegrid")
    sns.set_context({"figure.figsize": (24, 10)})

    p2 = sns.barplot(x = top_players[::-1], y = df_top_[::-1]/max(df_top_[::-1]), color = sns.diverging_palette(145, 280, s=85, l=25, n=7)[5],alpha = 0.5)
    p2.set_xticklabels(top_players[::-1],rotation=60)
    p2.set_title("%s-%s Top %s players based on our proposed metric" %(year,year+1,top),fontsize = 25)
    p2.set_ylabel("Relative value w.r.t maximum value obtained")
    sns.despine(left=True)
    for i in range(0,top,1):
        if floored[(top-1)-i]<100:
            p2.text(-0.16 + i, (df_top_[(top-1)-i]/max(df_top_))/2,str(floored[(top-1)-i]), fontsize = 20,color = 'white',alpha=1)
        else:
            p2.text(-0.23 + i, (df_top_[(top-1)-i]/max(df_top_))/2-0.07,str(floored[(top-1)-i]), fontsize = 20,color = 'white',alpha=1)
        

    p2.text(-0.16, 0.89, "61:  Absolute metric value floored to the nearest integer", fontsize = 15,color = 'black',alpha=1)
    return p2
```

```python
def metric_plot_playercentered(player,features,year,top=10,player_name = None):
    
    metric_ = features.iloc[23,:]
    order = np.argsort(metric_.values)[::-1]
    df_top_ = metric_[order]
    
    idx = np.where(df_top_.index.values==player)[0][0]
    
    df_top_ = df_top_[idx-top:idx+top]
    rank = [i+1 for i in range(idx-10,idx+10)[::-1]]
    xx = [0]*len(df_top_)
    xx[top-1] = df_top_[player]/max(df_top_[::-1])
    
    if player_name is None:
        top_players = [w.encode('raw_unicode_escape').decode('utf-8') for w in df_top_.index.values]
    else:
        top_players = player_name
        
    floored = [int(np.floor(d)) for d in df_top_]

    sns.set_style("whitegrid")
    sns.set_context({"figure.figsize": (24, 10)})
    
    sns.barplot(x = top_players[::-1], y = xx, color = 'red',alpha = 0.5)
    p2 = sns.barplot(x = top_players[::-1], y = df_top_[::-1]/max(df_top_[::-1]), color = sns.diverging_palette(145, 280, s=85, l=25, n=7)[5],alpha = 0.5)
    p2.set_xticklabels([str(rank[p])+": "+ top_players[::-1][p] for p in range(len(top_players[::-1]))],rotation=60)
    p2.set_title(player+"'s ranking in %s-%s based on our metric" %(year,year+1),fontsize = 25)
    p2.set_ylabel("Relative value w.r.t maximum value obtained")
    sns.despine(left=True)
    for i in range(0,2*top,1):
        if floored[(2*top-1)-i]<100:
            p2.text(-0.16 + i, (df_top_[(2*top-1)-i]/max(df_top_))/2,str(floored[(2*top-1)-i]), fontsize = 20,color = 'white',alpha=1)
        else:
            p2.text(-0.23 + i, (df_top_[(2*top-1)-i]/max(df_top_))/2-0.07,str(floored[(2*top-1)-i]), fontsize = 20,color = 'white',alpha=1)
        

    p2.text(-0.16, 0.89, "61:  Absolute metric value floored to the nearest integer", fontsize = 15,color = 'black',alpha=1)
    return p2
```

```python
def ranking_plot(features,year,top):
    plot_df = features.iloc[[4,11,22],:]
    plot_df = plot_df.append(plot_df.iloc[0,:]+plot_df.iloc[1,:],ignore_index=True).iloc[[2,3],:]
    
    ranking_classical = np.array(plot_df.iloc[1,:].values.astype(int))
    ranking_classical = np.argsort(ranking_classical)[-top:][::-1]
    ranking_metric = np.array(plot_df.iloc[0,:].values.astype(int))
    ranking_metric = np.argsort(ranking_metric)[-top:][::-1]
    
    d1 = dict(zip(ranking_classical,range(1,1+len(ranking_classical))))
    d2 = dict(zip(ranking_metric,range(1,1+len(ranking_metric))))
    
    toplot = []
    for i in d1.keys():
        if i in d2.keys():
            toplot.append(d2[i])
        else:
            toplot.append(100000)
    
    fig = plt.figure()                                                               
    ax = fig.add_subplot(1,1,1)  
    plt.axes().set_aspect('equal')
    ax.scatter(d1.values(),toplot,color = 'black',alpha=1)
    ax.plot(range(top+2),'r--',color = sns.color_palette("RdBu_r", 7)[6],alpha = 1,linewidth=1)
    ax.set_ylim(0,top+1)
    ax.set_xlim(0,top+1)
    ax.set_ylabel("Based on proposed Metric")
    ax.set_xlabel("Based on traditional Statistics")
    ax.set_title("Differences in Rankings for the %s-%s" % (year,year+1))

    major_ticks = np.arange(0, top+1, 5)                                              
    minor_ticks = np.arange(0, top+1, 1)                                               

    ax.set_xticks(major_ticks)                                                       
    ax.set_xticks(minor_ticks, minor=True)   
    ax.tick_params(axis='x', labelsize=11)
    ax.tick_params(axis='y', labelsize=11)
    ax.set_yticks(major_ticks)                                                       
    ax.set_yticks(minor_ticks, minor=True)                                           

    ax.grid(which="both")
    ax.grid(which='minor', alpha=0.02,color='blue')                                                
    ax.grid(which='major', alpha=0.07,color='blue')   
```

```python
def feature_plot(features_frame,year,features_wanted,players,players_nospecial = None):
    trans_dic = {'avg_value':'Average score','avg_value_smoothed': 'Smoothed average score', 'nassists': 'Assists', 'naway_goals': 'Away goals','nearly_assists': 'Early assists','nearly_goals': 'Early goals','nequalizer_assists': 'Assists on equalizer','nequalizer_goals': 'Equalizer goals','ngoals': 'Goals','nhome_assists': 'Home assists', 'nhome_goals': 'Home goals','nlate_assists': 'Late assists','nlate_goals': 'Late goals','nmid_assists': 'Assits in heart of game','nsub_assists': 'Assists when subbed in', 'nsub_goals': 'Goals when subbed in','ntiebreaker_assists': 'Tiebreaker assists','ntiebreaker_goals': 'Tiebreaker goals','season_value':'Overall score'}
    df = features_frame.loc[features_wanted,[p for p in players]]
    feat = [j for i in df.values for j in i]
    if players_nospecial is None:
        players_names = players * len(features_wanted)
    else:
        players_names = players_nospecial * len(features_wanted)
    features_names = [[trans_dic[features_wanted[i]] for j in range(len(players))] for i in range(len(features_wanted))]
    features_names = sum(features_names, [])
    dic = {'feature': pd.Series(feat),' ': pd.Series(players_names),'Features': pd.Series(features_names)}
    df_touse = pd.DataFrame(dic)
    
    sns.set_style("whitegrid")
    sns.set_context({"figure.figsize": (24, 24)})
    
    ax = sns.barplot(x="Features", y="feature", hue=" ", data=pd.DataFrame(df_touse), color=sns.color_palette("Blues",3)[1])
    if len(players) == 1:
        if players_nospecial is None:
            plt.title("%s's %s-%s season" % (players[0],year,year+1),size=30)
        else:
            plt.title("%s's %s-%s season" % (players_nospecial[0],year,year+1),size=30)
    else:
        if players_nospecial is None:
            plt.title("%s's and %s's %s-%s season in comparison" % (players[0],players[1],year,year+1),size=30)
        else:
            plt.title("%s's and %s's %s-%s season in comparison" % (players_nospecial[0],players_nospecial[1],year,year+1),size=30)
    ax.set_yticks(range(0,max(feat)+5,5))
    plt.ylabel(' ')
    plt.xlabel(' ')
    plt.ylim(0,max(feat)+5)
```
