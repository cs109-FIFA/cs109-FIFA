# Plot 3:

```python
plot3_df = features12.iloc[[4,11,22],:]
```

```python
plot3_df = plot3_df.append(plot3_df.iloc[0,:]+plot3_df.iloc[1,:],ignore_index=True)
```

```python
plot3_df = plot3_df.iloc[[2,3],:]
```

```python
ranking_classical = np.array(plot3_df.iloc[1,:].values.astype(int))
ranking_classical = np.argsort(ranking_classical)[-100:][::-1]
ranking_metric = np.array(plot3_df.iloc[0,:].values.astype(int))
ranking_metric = np.argsort(ranking_metric)[-100:][::-1]
```

```python
d1 = dict(zip(ranking_classical,range(1,1+len(ranking_classical))))
d2 = dict(zip(ranking_metric,range(1,1+len(ranking_metric))))
```

```python
toplot = []
for i in d1.keys():
    if i in d2.keys():
        toplot.append(d2[i])
    else:
        toplot.append(100000)
```

```python
fig = plt.figure()                                                               
ax = fig.add_subplot(1,1,1)  
plt.axes().set_aspect('equal')
ax.scatter(d1.values(),toplot,color = 'black',alpha=1)
ax.plot(range(101),'r--',color = sns.diverging_palette(10, 220, sep=80, n=7)[0],alpha = 0.5 )
ax.set_ylim(0,100)
ax.set_xlim(0,100)
ax.set_ylabel("Based on proposed Metric")
ax.set_xlabel("Based on traditional Statistics")
ax.set_title("Differences in Rankings")

major_ticks = np.arange(0, 101, 5)                                              
minor_ticks = np.arange(0, 101, 1)                                               

ax.set_xticks(major_ticks)                                                       
ax.set_xticks(minor_ticks, minor=True)   
ax.tick_params(axis='x', labelsize=11)
ax.tick_params(axis='y', labelsize=11)
ax.set_yticks(major_ticks)                                                       
ax.set_yticks(minor_ticks, minor=True)                                           

ax.grid(which="both")
ax.grid(which='minor', alpha=0.2)                                                
ax.grid(which='major', alpha=0.5)   
```
