## The following function will parse the data scraped above to extract features for the machine learning algorithms that we'll use to predict player value

```python
"""
Function
--------
extract_features receives a player data frame with match statistics and extracts the features 
                to be used for classificaton and prediction training.

Inputs
------
playerdf: a dataframe keyed by player and containing rows for each individual match day.
     For each player, match day pair there is a a dictionary with the following fields:
     
     goals: list of goals scored by that player (minute, was tiebreaker, was equalizer)
     assists: list of assists created by that player (minute, was tiebreaker, was equalizer)
     was_sub: Whether or not the player was a substitute
     was_home: Whether or not the player was on the home team or not
     match_value: Calculated value (from player_val) for that player from the specified match
     avg_val_smoothed: Avg. value of the player with a five match moving filter
     cum_val: Cumulative value the player has had to that point in the season.
   
Returns
-------
 a dataframe, keyed by player name with features extracted from playerdf that records
              the number of goals and assists the player contributed throughout the season
              with relevant attributes (time, home/away, equalizer, tiebreaker) also registered

Notes
-----
Any additional features may be included at a later date.
Written 24 Nov. 2015 by TWK
"""

def extract_features(playerdf):
    playerlist = playerdf.columns
    
    featuredict = {} # Empty dictionary that will be keyed by each player and will contain a dictionary of their features
    
    for player in playerlist:

        cum_value, smoothed_value, match_value = [], [], []
        season_value, avg_value, avg_value_smoothed = 0, 0, 0
        nhome_goals, nhome_assists = 0, 0
        naway_goals, naway_assists = 0, 0
        nequalizer_goals, nequalizer_assists = 0, 0
        ntiebreaker_goals, ntiebreaker_assists = 0, 0
        ngoals, nassists = 0,0
        nsub_goals, nsub_assists = 0, 0
        nearly_goals, nearly_assists = 0, 0
        nmid_goals, nmid_assists = 0, 0
        nlate_goals, nlate_assists = 0, 0
        
        pstats = playerdf[player]
        
        for match in np.arange(1,39):
            cum_value.append(pstats[match]['cum_val'])
            smoothed_value.append(pstats[match]['avg_val_smooth'])
            match_value.append(pstats[match]['match_value'])
            
            # Check if player recorded a statistic
            if isinstance(pstats[match]['was_home'],bool):  # Filters out default value
                for goal in pstats[match]['goals']:
                    if goal[0] > 0:
                        
                        # Check on home/away and increment
                        if pstats[match]['was_home']: nhome_goals += 1
                        else: naway_goals += 1
                        
                        # Check the game state of the goal and increment
                        if goal[1]: ntiebreaker_goals += 1
                        if goal[2]:nequalizer_goals += 1
                        
                        # Check on time aspect of the goal and increment
                        if goal[0] <= 20: nearly_goals += 1
                        elif goal[0] > 80: nlate_goals += 1
                        else: nmid_goals += 1
                        
                        # Check whether player was a substitute and increment
                        if pstats[match]['was_sub']: nsub_goals += 1
                            
                        # Simply increment the total number of goals.
                        ngoals += 1

                for assist in pstats[match]['assists']:
                    if assist[0] > 0:   # Just making sure (I don't think that assists are ever credited on own goals) 
                        # Check on home/away and increment
                        if pstats[match]['was_home']: nhome_assists += 1
                        else: naway_assists += 1
                        
                        # Check the game state of the goal and increment
                        if assist[1]: ntiebreaker_assists += 1
                        if assist[2]:nequalizer_assists += 1
                        
                        # Check on time aspect of the goal and increment
                        if assist[0] <= 20: nearly_assists += 1
                        elif assist[0] > 80: nlate_assists += 1
                        else: nmid_assists += 1
                        
                        # Check whether player was a substitute and increment
                        if pstats[match]['was_sub']: nsub_assists += 1
                            
                        # Simply increment the total number of assists.
                        nassists += 1
        season_value = cum_value[-1]
        avg_value = np.mean(match_value)
        avg_value_smoothed = np.mean(smoothed_value)
        
        featuredict[player] = {'nhome_goals': nhome_goals, 'naway_goals': naway_goals, 
                               'nequalizer_goals': nequalizer_goals, 'ntiebreaker_goals': ntiebreaker_goals, 
                               'ngoals': ngoals, 'nsub_goals': nsub_goals, 'nearly_goals': nearly_goals, 
                               'nmid_goals': nmid_goals, 'nlate_goals': nlate_goals, 
                               'nhome_assists': nhome_assists, 'naway_assists': naway_assists, 
                               'nequalizer_assists': nequalizer_assists, 'ntiebreaker_assists': ntiebreaker_assists, 
                               'nassists': nassists, 'nsub_assists': nsub_assists, 'nearly_assists': nearly_assists, 
                               'nmid_assists': nmid_assists, 'nlate_assists': nlate_assists, 'avg_value': avg_value, 
                               'avg_value_smoothed': avg_value_smoothed, 'season_value': season_value, 
                               'cum_value_list': cum_value, 'smoothed_value_list': smoothed_value,
                               'match_value_list': match_value}
        
    return pd.DataFrame(featuredict)
        
```

## Extract the features into new dataframes, pickle them for later use

```python
%time
features12 = extract_features(players12)
features13 = extract_features(players13)
features12.to_pickle('features12-13.pkl')
features13.to_pickle('features13-14.pkl')
```

## Combine the features into the match statistics frames and re-pickle them for any later use.

```python
frames12 = [players12, features12]
frames13 = [players13, features13]
combined12 = pd.concat(frames12)
combined13 = pd.concat(frames13)
```
